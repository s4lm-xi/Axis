```python
import tkinter as tk
import random
import time

# Initialize variables
sentence = "hello i am axis i can transcibe sign language to speech"
generated_text = ""

# Function to generate gibberish text and original sentence
def generate_gibberish():
    
    
    global generated_text
    final_sentence = ""

    for letter in sentence:
        gibberish = ""
        for i in range(3):
            gibberish += f"pin{random.randint(1, 5)} + pin{random.randint(1, 5)} "
            text_var.set(gibberish)
            window.update_idletasks()
            time.sleep(0.1)
        
        generated_text += gibberish
        final_sentence += letter
        final_sentence_label.config(text=final_sentence)
        generated_text_label.config(text=generated_text)
        window.update_idletasks()
        time.sleep(0.5)
        
    text_var.set('')

# Create the main window
window = tk.Tk()
window.title("Axis logs")

# Set the window size to a fixed value (width x height)
window.geometry("850x200")  # You can adjust the width and height as needed

# Create a label to display the generated gibberish
text_var = tk.StringVar()
generated_text_label = tk.Label(window, textvariable=text_var, font=("Helvetica", 12))
generated_text_label.pack(pady=20)

# Create a label to display the final sentence
final_sentence_label = tk.Label(window, text="", font=("Helvetica", 25))
final_sentence_label.pack(pady=20)

generate_gibberish()

# Run the GUI
window.mainloop()
```
