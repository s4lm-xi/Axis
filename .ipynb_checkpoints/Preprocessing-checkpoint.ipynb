```python
import pandas as pd
import json
```

```python
data = pd.read_csv('letter_combination.csv')
```

```python
letters = data['letters'].to_list()
hand_1 =  data['hand_1'].to_list()
hand_2 = data['hand_2'].to_list()


hand_1 = [[int(j) for j in i.split(',')] for i in hand_1]
hand_2 = [[int(j) for j in i.split(',')] for i in hand_2]
```

```python
payload = {}

for idx in range(len(letters)):
    payload[letters[idx]] = {
        'hand_1': hand_1[idx],
        'hand_2': hand_2[idx]
    }
```

```python
with open('letter_combination.json', 'w') as file:
    file.write(json.dumps(payload))
```
